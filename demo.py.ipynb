```python
import openai
from dotenv import load_dotenv, find_dotenv
import os
from sklearn.metrics.pairwise import cosine_similarity

from openai import OpenAI

load_dotenv(find_dotenv())
```

```python
client = OpenAI()
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a helpful assistant, give short concise answers"},
        {"role": "user", "content": "What is the answer to the universe?"}
    ]
)
response.choices[0].message.content
```

```python

def get_closest(query: str, contents: [str]):
    def get_embedding(c: str):
        return client.embeddings.create(input=c, model="text-embedding-3-large").data[0].embedding

    query_embedding = get_embedding(query)
    contents_embeddings = [get_embedding(content) for
                           content in contents]
    best_similarity = 0.0
    nearest_content_index = 0
    for i in range(len(contents)):
        similarity = cosine_similarity([query_embedding], [contents_embeddings[i]])[0][0]
        print(f"{contents[i]}: {similarity}")
        if similarity > best_similarity:
            best_similarity = similarity
            nearest_content_index = i
    return contents[nearest_content_index]
```

```python
get_closest("PC", ["Fernseher", "Computer", "Mixer", "Thermoskanne", "Sessel"])
```
